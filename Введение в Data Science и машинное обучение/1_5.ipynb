**1_5_1**
У какой доли студентов из датасэта в колонке lunch указано free/reduced?

```python
students_perfromance.loc[students_perfromance.lunch == 'free/reduced'].shape[0]/students_perfromance.shape[0]
```

**1_5_2** Как различается среднее и дисперсия оценок по предметам у групп студентов со стандартным или урезанным ланчем?

```python
students_perfromance.groupby('lunch').aggregate({'math score':['mean', 'var'], 'reading score':['mean', 'var'], 'writing score' :['mean', 'var']})
```
